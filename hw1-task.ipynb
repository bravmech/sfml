```python
# (6) Создать новый строковый столбец в data frame-е, который:
# принимает значение A, если значение LIMIT_BAL <=10000
# принимает значение B, если значение LIMIT_BAL <=100000 и >10000
# принимает значение C, если значение LIMIT_BAL <=200000 и >100000
# принимает значение D, если значение LIMIT_BAL <=400000 и >200000
# принимает значение E, если значение LIMIT_BAL <=700000 и >400000
# принимает значение F, если значение LIMIT_BAL >700000

def limit_step(limit):
    if limit <= 10000:
        return 'A'
    elif limit <= 100000:
        return 'B'
    elif limit <= 200000:
        return 'C'
    elif limit <= 400000:
        return 'D'
    elif limit <= 700000:
        return 'E'
    else:
        return 'F'

df['LIMIT_STEP'] = df['LIMIT_BAL'].map(lambda x: limit_step(x))
df.head()
```

```python
def pairplot_wrapper(df, x_vars, y_vars, hue):
    plt.figure()
#     print('x_vars = {}'.format(', '.join(x_vars)))
#     print('y_vars = {}'.format(', '.join(y_vars)))
    sns.pairplot(df, x_vars=x_vars, y_vars=y_vars, hue=hue)
```

```python
df = df.drop('LIMIT_STEP', errors='ignore')
columns = df.columns[:-1]
shape1 = len(columns) # 24
n = shape1 // 4
chunks = list(zip(*[iter(columns)] * n))
print(chunks)
```

```python
for i in range(len(chunks)):
    for j in range(i, len(chunks)):
        pairplot_wrapper(df, chunks[i], chunks[j], hue='default.payment.next.month')
```
